## 1. Tokenization

Tokenization in NLP is like slicing a loaf of bread: it breaks down text into smaller, manageable pieces called tokens. These tokens can be words, phrases, or even characters.

image.png

## Implementation

```python
import nltk
from nltk import word_tokenize
```

```python
nltk.download('punkt')
```

```python
sentence = """An apple is a round, edible fruit produced by an apple
tree (Malus spp., among them the domestic or orchard apple;
Malus domestica). Apple trees are cultivated worldwide and are
the most widely grown species in the genus Malus."""
```

```python
tokens = word_tokenize(sentence)
```

```python
tokens
```

## 2. Stop words

Stop words in NLP are the most common words in a language that are often ignored during text processing because they carry less meaningful information. These words include articles, prepositions, conjunctions, etc. Examples include "is", "and", "the", "in", "on", and many more.

image.png

**Why Exclude Stop Words?**
1. **Reduces Noise:** Eliminating stop words helps to focus on the important words that carry significant meaning.

2. **Improves Efficiency:** Fewer words lead to faster processing and reduced computational load.

3. **Enhances Accuracy:** Helps in improving the performance of text analysis tasks like classification, clustering, and information retrieval.

## Implementations

```python
from nltk.corpus import stopwords
```

```python
nltk.download("stopwords")
```

```python
stop_words = set(stopwords.words("english"))
```

```python
stop_words
```

```python
filtered_words = [word for word in tokens if word.lower() not in stop_words]
```

```python
filtered_words
```

## 3. Stemming

Stemming in NLP reduces words to their base or root form. For example, "running", "runner", and "ran" all become "run". This helps simplify text and reduce it to core meanings. There are different stemming algorithms, but one of the most common is the Porter Stemmer.

image.png

**Why Stemming?**
1. **Reduces Complexity:** Simplifies the text, making it easier to analyze.

2. **Improves Accuracy:** Helps in matching similar words and reduces redundancy.

3. **Efficiency:** Lowers the number of unique terms in text data, saving on computational resources.

## Implementation

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
stemmed_words = [stemmer.stem(word) for word in filtered_words]
```

```python
stemmed_words
```

## 4. Lemitization

Lemmatization in NLP (Natural Language Processing) is the process of reducing words to their base or root form, but unlike stemming, it ensures that the resulting words are actual words. It's context-aware and takes into account the meaning of the words, leading to more accurate results.

image.png

**Why Lemmatization?**
1. **Precision:** Produces actual words, not just truncated forms.

2. **Context-Aware:** Takes into account the part of speech and meaning.

3. **Improves Text** Analysis: Helps in better matching and understanding of words.

## implementations

```python
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
```

```python
lemmatized_words
```
